## 0_[umap-learn](https://umap.scikit-tda.org/)

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from umap import UMAP

# 生成随机数据
data = np.random.rand(100, 10)

data
```

![](../img/umap_1.png)

```python
"""
使用UMAP进行降维时

n_neighbors：指定高维空间中每个数据点的相邻点个数，控制umap平衡全局结构和局部结构
- 参数值越大：丢失越多的局部结构，保留越多的全局结构
- 参数值越小：丢失越多的全局结构，保留越多的局部结构

min_dist:控制空间中点之间的最小距离，控制umap将点聚集在一起的紧密程度
- 参数值越小：嵌入紧密
- 参数值越大：嵌入松散

metric：计算高维空间中点的距离

"""
umap_model = UMAP(n_neighbors=5, min_dist=0.3, metric='euclidean')


embedding = umap_model.fit_transform(data)

embedding
```

```python
# 绘制散点图
plt.scatter(embedding[:, 0], embedding[:, 1])
plt.show()
```
